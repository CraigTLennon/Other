```python
#Matrix multiplication is dot .
# * is elementwise mult
A=np.array([[1,2],[3,4]])
B=np.array([[1,.3],[2,.1]])
A*B
A.dot(B)
#invers is 
np.linalg.inv(A)
np.linalg.inv(A).dot(A)
```

```python
#outer product, used for covariance
C=np.array([1,2,3,4,5])
np.outer(C,C)
X=np.random.randn(100,3)
cov=np.cov(X.T)
cov
```

```python
#eigenvalues, eigenvectors
np.linalg.eig(cov) #eigenvalues and eigenvectors
```

```python
#solving linear systems Ax=b (assume A intertivble)
b=np.array([1,2])
x=np.linalg.inv(A).dot(b)
x
# or 
np.linalg.solve(A,b) #use this andnot inverse
```

```python
#general gaussian, start with covar mtx
cov=np.array([[1,.8],[.8,2]])
from scipy.stats import multivariate_normal
y=multivariate_normal.rvs(mean=np.array([0,2]),cov=cov,size=1000)
```
