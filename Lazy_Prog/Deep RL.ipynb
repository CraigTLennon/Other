#### Deep RL
RBF Neural networks:  RBF is a radial basis function, and RBF networks can be thought of as linear models with feature extraction in which the features is a RBF kernel, or a 1-hidden layer NN with RBF kernel as activation function. An RBF is a non-normalized gaussian, and it depends on the distance between the vectors, not direction, which is what makes is radial.
$$
\phi(x) = \exp \left( -\frac{|x-c|^2}{\sigma^2} \right)
$$
The number of basis functions to choose is a hyperparameter, and the points c are called exemplars.  You could just sample the space to get some options.  Scikitlearn has an RBF sampler, and other RBF methods.  The features do not change as we learn.  Sci kit learn also has an SGD method.
